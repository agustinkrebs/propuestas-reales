```python
import requests
import pandas as pd
import json
import os
```

```python
URL = "http://www.propuestasreales.cl/api/posts"
r = requests.get(url = URL).json()
data = r['data']
data_attributes = map(lambda x : x['attributes'], data)
df = pd.DataFrame.from_dict(data_attributes, orient='columns')
```

```python
df.to_csv(os.path.join("respuestas", 'respuestas_API.csv'))
```
