```python
import pandas as pd
import os 
```

```python
df = pd.read_csv("respuestas" + os.sep + "respuestas_v2.csv", sep=',')
```

```python
df.columns = ["fecha", "caracter", "correo", "instagram", "ministerio", "tipo", "descripcion", "comentario_extra"]
```

```python
ministerios_id = {
    'Economía, Fomento y Turismo': 1,
    'Salud': 2,
    'Mujer y la Equidad de Género': 3,
    'Hacienda': 4,
    'Desarrollo Social y Familia': 5,
    'Trabajo y Previsión Social': 6,
    'Vivienda y Urbanismo': 7,
    'Transportes y Telecomunicaciones': 8,
    'Defensa Nacional': 9,
    'Educación': 10,
    'Obras Públicas': 11,
    'Ciencia, Tecnología, Conocimiento e Innovación': 12,
    'Interior y Seguridad Pública': 13,
    'Secretaría General de Gobierno': 14,
    'Justicia y Derechos Humanos': 15,
    'Minería': 16,
    'Energía': 17,
    'Relaciones Exteriores': 18,
    'Medio Ambiente': 19,
    'Culturas, las Artes y el Patrimonio': 20,
    'Secretaría General de la Presidencia': 21,
    'Agricultura': 22,
    'Bienes Nacionales': 23,
    'Deporte': 23}

def lista_de_id(ministerios):
    ids = []
    for ministerio in ministerios_id.keys():
        if ministerio in ministerios:
            ids.append(ministerios_id[ministerio])
    return ids
```

```python
print("const ministriesPostData = [")
i = 0
for dato in df.itertuples():
    i += 1
    id_ministerios = lista_de_id(dato.ministerio)
    for min_id in id_ministerios:
        print('{')
        print(f"""
        ministryId: {min_id},
        postId: {i},
        createdAt: new Date(),
        updatedAt: new Date(),""")
        print('},')
        
print("];")
```
